```python
data = pd.read_csv('drive/My Drive/Colab/13e6.csv', sep=',', na_values=".")
data
```

```python
aa=data[['a']].to_numpy()
bb=data[['b']].to_numpy()
cc=data[['c']].to_numpy()
dd=data[['d']].to_numpy()
ab=aa*bb
bc=bb*cc
ac=aa*cc
ad=aa*dd
bd=bb*dd
cd=cc*dd
abc=aa*bb*cc
abd=aa*bb*dd
bcd=bb*cc*dd
acd=aa*cc*dd
abcd=aa*bb*cc*dd
```

```python
data['ab']=ab
data['bc']=bc
data['ac']=ac
data['ad']=ad
data['bd']=bd
data['cd']=cd
data['abc']=abc
data['abd']=abd
data['acd']=acd
data['bcd']=bcd
data['abcd']=abcd
X=data[['a', 'b', 'c', 'd', 'ab', 'bc', 'ac', 'ad', 'bd', 'cd', 'abc', 'abd', 'bcd', 'acd', 'abcd']]
X = sm.add_constant(X) # adding a constant
X
```

```python
Y=data[['Replicate']]
Y
```
